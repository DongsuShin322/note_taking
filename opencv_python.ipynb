## 제 6강 : 이미지 임계처리

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('bookpage.jpg')
ret1, threshold1 = cv2.threshold(img, 12, 255, cv2.THRESH_BINARY)

grayscaled = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret2, threshold2 = cv2.threshold(grayscaled, 12, 255, cv2.THRESH_BINARY)

gaus = cv2.adaptiveThreshold(grayscaled, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
```

### Rotation : 이미지 회전
### cv2.getRotationMatrix2D(이미지 중심 좌표, 회전 각도, 스케일)

```python
img = cv2.imread('python.jpeg')
rows, cols = img.shape[:2]
```

### Affine Transformation : 평행선 유지, 이미지 변환

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('chess.png')
rows, cols = img.shape[:2]
```

```python
pts1 = np.float32([[20,10],[40,10],[20,20]])
pts2 = np.float32([[20,30],[40,20],[20,40]])
```

```python
# Affine 변환을 위해서는 3개의 Match가 되는 점이 있으면 변환행렬을 구할 수 있음
M = cv2.getAffineTransform(pts1, pts2)
dst = cv2.warpAffine(img, M, (cols,rows))
```

```python
plt.subplot(121),plt.imshow(img),plt.title('image')
plt.subplot(122),plt.imshow(dst),plt.title('Affine')
```

### Perspective Transformation : 직선 유지, 평행 유지 안됨

```python
img = cv2.imread('railload.jpeg')
rows, cols = img.shape[:2]
```

```python
# 포인트 순서 : 좌상->좌하->우상->우하
pts1 = np.float32([[150,57],[110,100],[222,57],[262,100]])
pts2 = np.float32([[10,10],[10,1000],[1000,10],[1000,1000]])
```

```python
# 4개의 Point의 Input값과이동할 output Point 가 필요
M = cv2.getPerspectiveTransform(pts1, pts2)
dst = cv2.warpPerspective(img, M, (1100,1100))
```

```python
plt.subplot(121),plt.imshow(img),plt.title('image')
plt.subplot(122),plt.imshow(dst),plt.title('Perspective')
```

## 이미지 윤곽선

### cv2.findContours(이미지, 윤곽선 찾는 방법, 윤곽선 찾을 때 근사치 방법)
### returns : image, contours , hierachy

### cv2.drawContours(이미지, 윤곽선 정보, 윤곽선 인덱스, 색, 두께)

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('rectangles.png')
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# grayscale -> binary image
ret, thresh = cv2.threshold(imgray, 127, 255, 0)
```

```python
image, contours, hierachy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
image = cv2.drawContours(img, contours, -1, (0,255,0), 3)
```

```python
cv2.imshow('image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
