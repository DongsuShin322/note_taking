### url로 file 다운로드 받아 저장하기

```python
import urllib.request

url = 'http://uta.pw/shodou/img/28/214.png'
savename = "/home/dongsu/test.png"

urllib.request.urlretrieve(url, savename)
print('saved...') 
```

```python
url = 'http://uta.pw/shodou/img/28/214.png'
savename = "/home/dongsu/test.png"

mem = urllib.request.urlopen(url).read() # 메모리에 올려 놓은 후

with open(savename, 'wb') as f:
    f.write(mem)
    print ('saved...')
```

```python
url = 'http://api.aoikujira.com/ip/ini'
res = urllib.request.urlopen(url)
data = res.read() # 바이너리 파일

text = data.decode('utf-8') # 디코딩
print (text)
```

```python
import urllib.parse
API = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
values = {
    'stnId': '108' # 번호를 sys.argv[1]를 사용하여 사용자로부터 입력으로 받아 사용할 수도 있다 --- $ python this.py 108  
}
params = urllib.parse.urlencode(values) # 매개변수를 URL 인코딩
url = API + "?" + params
print ('url=',url)

data = urllib.request.urlopen(url).read()
text = data.decode('utf-8')
print (text)
```

```python
from bs4 import BeautifulSoup as bs
html = """
<html><body>
    <h1>스크레이핑이란?</h1>
    <p>웹 페이지를 분석하는 것</p>
    <p>원하는 부분을 추출하는 것</p>
</body></html>
"""
soup = bs(html, "html.parser")
h1 = soup.html.body.h1
p1 = soup.html.body.p
p2 = p1.next_sibling.next_sibling # Enter도 노드이므로 next_sibling 두 번

print('h1 = ' + h1.string)
print('p1 = ' + p1.string)
print('p2 = ' + p2.string)
```

```python
html = """
<html><body>
    <u1>
        <li><a href="http://www.naver.com">naver</a><li>
        <li><a href="http://www.daum.net">daum</a><li>
    </u1>
</body></html>    
"""
soup = bs(html,"html.parser")
links = soup.find_all('a')
for a in links:
    href = a.attrs['href']
    text = a.string
    print (text, ">", href)
```

```python
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
res = urllib.request.urlopen(url)
soup = bs(res, "html.parser")
title = soup.find("title").string
wf = soup.find("wf").string
print (title)
print (wf)
```

```python
html = """
<html><body>
<div id="meigen">
    <h1>위키북스 도서</h1>
    <ul class="items">
        <li>유니티 게임 이벤트 입문</li>
        <li>스위프트로 시작하는 아이폰 앱 개발 교과서</li>
        <li>모던 웹사이트 디자인의 정석</li>
    </ul>
</div>
</body></html>
"""
soup = bs(html, "html.parser")
h1 = soup.select_one("div#meigen > h1").string
print ("h1 =", h1)
li_list = soup.select("div#meigen > ul.items > li") # css 문법으로 #은 id를 표시
for li in li_list:
    print ("li =", li.string)
```

```python
url = "http://info.finance.naver.com/marketindex/"
res = urllib.request.urlopen(url)
soup = bs(res, "html.parser")
price = soup.select_one('div.head_info > span.value').string
print ('usd/krw =',price)
```

### CSS 선택자 알아보기

```python
html = """
<html><body>
 <ul id='bible'>
    <li id='ge'>Genesis</li>
    <li id='ex'>Exodus</li>
    <li id='le'>Leviticus</li>
    <li id='nu'>Numbers</li>
    <li id='de'>Deuteronomy</li>
 </ul>
</body></html>
"""
from bs4 import BeautifulSoup as bs
soup = bs(html, "html.parser")
```

```python
sel = lambda q : print (soup.select_one(q).string)
```

```python
# Numbers 요수 추출하기
sel('#nu') # id=nu 
sel('li#nu') # <li>태그에 id=nu
sel('ul > li#nu') # <ul>태그 자식인 <li>태그에 id=nu
sel('#bible #nu') # #bible 아래의 #nu
sel('#bible > #nu') # >:직접적인 부모-자식 관계
sel('ul#bible > li#nu') # id=bible인 <ul>태그 바로 아래에 있는 id=nu인 <li>태그
sel("li[id='nu']") # id=nu인 <li>태그
sel('li:nth-of-type(4)') # 4번째 <li>태그
print (soup.select('li')[3].string) 
print (soup.find_all('li')[3].string)
```

```python
html="""
<html><body>
<div id='main-goods' role='page'>
    <h1>과일과 야채</h1>
    <ul id='fr-list'>
        <li class='red green' data-lo='ko'>사과</li>
        <li class='purple' data-lo='us'>포도</li>
        <li class='yellow' data-lo='us'>레몬</li>
        <li class='yellow' data-lo='ko'>오렌지</li>
    </ul>
    <ul id='ve-list'>
        <li class='white green' data-lo='ko'>무</li>
        <li class='red green' data-lo='us'>파프리카</li>
        <li class='black' data-lo='ko'>가지</li>
        <li class='black' data-lo='us'>아보카도</li>
        <li class='white' data-lo='cn'>연근</li>
    </ul>
</div>
</body></html>    
"""
soup = bs(html, "html.parser")
```

```python
# 아보카도 추출하기
print (soup.select_one('li:nth-of-type(8)').string) # 모든 <li>태그 중 8번째 요소
print (soup.select_one('#ve-list > li:nth-of-type(4)').string) # id=ve-list 바로 아래에 있는 모든 <li>태그 중 4번째 요소
print (soup.select("#ve-list > li[data-lo='us']")[1].string) 
print (soup.select("#ve-list > li.black")[1].string)
```

```python
cond = {'data-lo':'us', 'class':'black'} # 딕셔너리 사용!
print (soup.find("li", cond).string)
print (soup.find(id='ve-list').find("li",cond).string)
```
